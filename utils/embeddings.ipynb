```python
from deepface import DeepFace
import os
```

```python
def create_embeddings(p):
    res = DeepFace.represent(img_path = p,model_name="ArcFace",enforce_detection = False,detector_backend='retinaface')
    vector = res[0]['embedding']
    return vector
```

```python
create_embeddings("../images/keyur/all/IMG-20260104-WA0009.jpg")
```

```python
Embeddings = {}
persons = "../images"
for person in os.listdir(persons):

    person_path = f"{persons}/{person}/all"
    Embeddings[person] = []
    for img in os.listdir(person_path):
        img_path = os.path.join(person_path,img)
        print(img_path)
        embed = create_embeddings(img_path)
        Embeddings[person].append(embed)
```

```python
print(Embeddings)
```

```python
import pickle
with open("../face_embeddings.pkl","wb")as f:
    pickle.dump(Embeddings,f)
print("Face embeddings saved")
```
